# Importing DrillingInfo Lease List to Find Serial/Parcel Numbers

```python
import fuzzywuzzy
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
drillingInfoLeases = pd.read_csv('Source Files/DrillingInfo Lease Export/Leases Table.csv')
drillingInfoLeases.head()
drillingInfoLeases.info()
```

### Cleaning up DrillingInfo Spreadsheet by Dropping Duplicate records 

```python
drillingInfoLeases.drop_duplicates(subset = ["Record Number", "Area (Acres)"], inplace = True)

drillingInfoLeases
```

### Finding 100% matches in DrillingInfo Spreadsheet

```python
drillingInfoLeases[drillingInfoLeases["Record Number"].isin(LeasesNotInGIS["LEASE NO."])]
```

####  Taking Just Vol/Page (Sale Parcel #) and Record Number from matched records DrillingInfo Spreadsheet

```python
matchesDIandMasterSpread = drillingInfoLeases[drillingInfoLeases["Record Number"].isin(LeasesNotInGIS["LEASE NO."])][["Record Number", "Vol/Page"]]
matchesDIandMasterSpread
```

### Cleaning Parcel Number from Vol/Page 

```python
matchesDIandMasterSpread["Sale Parcel Number"] = matchesDIandMasterSpread["Vol/Page"].apply(lambda x: x.split("/")[0])
matchesDIandMasterSpread.drop(columns = ["Vol/Page"], inplace = True)
```

```python
### Merging/Joining LeasesNotInGIS with DrillingInfo matched table
```

```python
LeasesNotInGIS = pd.merge(left = LeasesNotInGIS, right = matchesDIandMasterSpread, how = "left", left_on="LEASE NO.", right_on="Record Number")

LeasesNotInGIS
```

### Fuzzy matching Sale Parcel Numbers Cleaned From DriillingInfo to ones found in GIS List

```python
gisMatches = []

for parcel in LeasesNotInGIS["Sale Parcel Number"].fillna('blank'):
    match = process.extractOne(parcel, gis["LEASE_NO"], score_cutoff=90)
    
    if type(match) != type(None):
        gisMatches.append(match[0])
    else:
        gisMatches.append("")
        
```

```python
gisMatches
len(gisMatches)
```

### Adding Fuzzy Matched GIS lease/parcel numbers to LeasesNotInGIS to Easily Identify Which Lease Number in Master Spreadsheet belongs to GIS lease/parcel numbers

```python
LeasesNotInGIS["Serial/Parcel in GIS"] = gisMatches
```

### Reordering DataFrame to See Lease Number in Spreadsheet vs. GIS Parcel Number

```python
LeasesNotInGIS = LeasesNotInGIS[['LEASE NO.', 'Serial/Parcel in GIS', 'PLOTTED', 'GRANTEE', 'SALE DATE',
       'EFFECTIVE DATE OF LEASE', 'EXPIRATION DATE', 'ACRES', 'COUNTY', 'ST',
       'STATUS/COMMENTS', 'EXPIRATION YEAR', 'TOWNSHIP', 'RANGE',
       'DESCRIPTION I', 'DESCRIPTION II', 'DESCRIPTION III', 'Record Number',
       'Sale Parcel Number'] ]

```

# Analyzing Leases Listed Not Issued in Master Spreadsheet

```python
#creating filtered data frame of master spreadsheet that are labeled not issued
NotIssued = spreadsheet[spreadsheet["STATUS/COMMENTS"].apply(lambda x: "issued" in str(x).lower())]
```

```python
def fuzzyMatchCustom(query, choices):
    '''
    This function will fuzzy matching to find matches for lease serial numbers in our master spreadsheet to BLM lR2000 
    database serial numbers
    
    '''
    
    #assigingn an 88 score cutoff to avoid false matches
    match = process.extractOne(query, choices, score_cutoff=88)
    if type(match) == type(None):
        return "No Match"
    else:
        return match[0]
    
```

```python
#finding matches and adding column
NotIssued["BLM LR2000 Fuzzy Match"] = NotIssued["LEASE NO."].apply(lambda x: fuzzyMatchCustom(x, lr2000["Serial Number Full"]))
```

### Merging Fuzzy Matched Serials from Spreadsheet to LR 2000 Records

```python
NotIssued = pd.merge(left = NotIssued, right = lr2000[["Serial Number Full", "Disposition"]], how = "left",
                    left_on="BLM LR2000 Fuzzy Match", right_on = "Serial Number Full")
```

```python
#reorganizing columns for better viewiing
NotIssued = NotIssued[['LEASE NO.', 'BLM LR2000 Fuzzy Match',   'Serial Number Full',  'PLOTTED', 'GRANTEE',
       'SALE DATE', 'EFFECTIVE DATE OF LEASE', 'EXPIRATION DATE', 'ACRES',
       'COUNTY', 'ST', 'STATUS/COMMENTS', 'Disposition', 'EXPIRATION YEAR', 'TOWNSHIP',
       'RANGE', 'DESCRIPTION I', 'DESCRIPTION II', 'DESCRIPTION III',
     ]]
```

```python
NotIssued.drop_duplicates(subset = "LEASE NO.")
```

### Cross Checking Serials with No Match with DrillingInfo

```python
noMatches = NotIssued[NotIssued["BLM LR2000 Fuzzy Match"] == "No Match"]

noMatches
```

#### Fuzzy matching Parcel Number with Vol/Page Column in DrillingInfo Export

```python
noMatches["DrillingInfo Parcel FuzzyMatch"] = noMatches["LEASE NO."].apply(lambda x: fuzzyMatchCustom(x, drillingInfoLeases["Vol/Page"]))

noMatches
```

#### Merging Record Number from DrillingInfo Column based on Fuzzy Matched Parcel Number 
*Record number will be close to LR2000 Serial Number if not Exact

```python
noMatches = pd.merge(left=noMatches, right=drillingInfoLeases[["Record Number", "Vol/Page"]], how = "left", left_on = "DrillingInfo Parcel FuzzyMatch", right_on = "Vol/Page" )

noMatches
```

#### Dropping Columns that are in NoIssued Dataframe That were introduced when using fuzzy match to LR2000 & merged with LR2000 fields

```python
noMatches.drop(columns = ["BLM LR2000 Fuzzy Match", "Disposition", "Serial Number Full"], inplace = True)
```

#### Performing Fuzzy match to LR2000 Serial numbers from DrillingInfo's Record Number

```python
noMatches["BLM LR2000 Fuzzy Match"] = noMatches["Record Number"].apply(lambda x: fuzzyMatchCustom(x, lr2000["Serial Number Full"]))

noMatches
```

#### Merging/Joining LR2000 fields based on fuzzy match of record number  from drillinginfo and serial number from lr 2000

```python
noMatches = pd.merge(left=noMatches, right = lr2000[["Serial Number Full", "Disposition"]], how = "left",
                    left_on="BLM LR2000 Fuzzy Match", right_on = "Serial Number Full")

noMatches
```

### Finalizing NotIssued Data Table Union Joining/concatenating NoMatches & Not Issued and Dropping Duplicate Lease Numbers That Occured by Splitting Them Apart

```python
FinalNotIssued = pd.concat([NotIssued, noMatches])
FinalNotIssued
FinalNotIssued.drop_duplicates(subset = "LEASE NO.", keep='last', inplace = True)
```

#### Since The Tables were not aligned, the tables were sorted alphabetically

```python
FinalNotIssued = FinalNotIssued[columnOrder]
```
